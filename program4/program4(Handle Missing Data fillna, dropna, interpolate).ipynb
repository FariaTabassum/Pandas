```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
df.drop(['Unnamed: 4', 'Unnamed: 5'], axis = 1)
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
df.drop(['Unnamed: 4', 'Unnamed: 5'], axis = 1)
```

```python
#Use fillna() method in dataframe
new_df = df.fillna(0)
new_df
```

```python
df.drop(['Unnamed: 4', 'Unnamed: 5'], axis = 1)
```

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
new_df = df.fillna(method="ffill")#forward fill
new_df
```

```python
new_df = df.fillna(method="bfill")#backward fill
new_df
```

```python
#"axis" parameter in fillna() method in dataframe
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
#"limit" parameter in fillna() method in dataframe
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
#interpolate() to do interpolation in dataframe
new_df = df.interpolate()
new_df
```

```python
#interpolate() method "time"
new_df = df.interpolate(method="time") 
new_df
```

```python
df.drop(['Unnamed: 4', 'Unnamed: 5'], axis = 1)
```

```python
#dropna() method Drop all the rows which has "na" in dataframe
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```
